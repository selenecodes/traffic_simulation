```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../Data/data.csv", names=["max_speed", "avg_speed"])
```

```python
data = data.groupby(by=['max_speed']).mean()
data["percent_of_max"] = (data["avg_speed"] / data.index) * 100
data
```

```python
plt.plot(data.index, data['avg_speed'], color = 'g', label = "Average Speeds")
plt.xlabel("Maximum road speed")
plt.ylabel("Mean actual speed")
plt.title("Mean vs Max speed")
plt.legend()
plt.show()
```

```python
plt.plot(data.index, data["percent_of_max"], color = 'g', label = "Average percentile speed")
plt.xlabel("Maximum road speed")
plt.ylabel("Utilized Speed Percentage")
plt.title("Percentage of max speed")
plt.ylim(0,100)
plt.legend()
plt.show()
```
