```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_10 = pd.read_csv("../Data/data_10.csv", names=["max_speed", "avg_speed"]).groupby(by=['max_speed']).mean()
data_20 = pd.read_csv("../Data/data_20.csv", names=["max_speed", "avg_speed"]).groupby(by=['max_speed']).mean()
data_40 = pd.read_csv("../Data/data_40.csv", names=["max_speed", "avg_speed"]).groupby(by=['max_speed']).mean()
data_60 = pd.read_csv("../Data/data_60.csv", names=["max_speed", "avg_speed"]).groupby(by=['max_speed']).mean()
```

```python
data_10["percent_of_max"] = (data_10["avg_speed"] / data_10.index) * 100
data_20["percent_of_max"] = (data_20["avg_speed"] / data_20.index) * 100
data_40["percent_of_max"] = (data_40["avg_speed"] / data_40.index) * 100
data_60["percent_of_max"] = (data_60["avg_speed"] / data_60.index) * 100
```

```python
plt.scatter(data_10.index, data_10["avg_speed"], alpha = .5, color = 'orange', label = "Average Speeds, 20 cars")
plt.scatter(data_20.index, data_20['avg_speed'], alpha = .5, color = 'r', label = "Average Speeds, 20 cars")
plt.scatter(data_40.index, data_40['avg_speed'], alpha = .5, color = 'g', label = "Average Speeds, 40 cars")
plt.scatter(data_60.index, data_60['avg_speed'], alpha = .5, color = 'b', label = "Average Speeds, 60 cars")

plt.xlabel("Maximum road speed")
plt.ylabel("Mean actual speed")
plt.title("Mean vs Max speed")
plt.legend()
plt.show()
```

```python
plt.scatter(data_10.index, data_10["percent_of_max"], alpha = .5, color = 'orange', label = "Average percentile speed, 10 cars")
plt.scatter(data_20.index, data_20["percent_of_max"], alpha = .5, color = 'r', label = "Average percentile speed, 20 cars")
plt.scatter(data_40.index, data_40["percent_of_max"], alpha = .5, color = 'g', label = "Average percentile speed, 40 cars")
plt.scatter(data_60.index, data_60["percent_of_max"], alpha = .5, color = 'b', label = "Average percentile speed, 60 cars")

plt.xlabel("Maximum road speed")
plt.ylabel("Utilized Speed Percentage")
plt.title("Percentage of max speed")
plt.ylim(75,100)
plt.legend()
plt.show()
```
